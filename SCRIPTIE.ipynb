```python
def removeNone(lis):
    result = []
    for i in lis:
        if i != 'NONE' and i != 'NONE;;;;;;;;;;;;':
            result.append(i)
    return result

current = file3.copy()

current['PROTEINVALUE'] = 'onbepaald'
for index,val in enumerate(file3.itertuples()): 
    ingredient = file3.loc[index,'INGREDIENT'] 
    subs = [i for i in df2[df2['INGREDIENT'] == ingredient].iloc[0]]
    subs = removeNone(subs)
    #print(subs)
    max_value = float(df1[df1['INGREDIENT'] == ingredient]['PROTEINVALUE'])
    max_ingredient = ingredient
    for i in subs:
        if (i in df1['INGREDIENT'].unique()):
            value = float(df1[df1['INGREDIENT'] == i]['PROTEINVALUE'])
            #print(value)
            if value > max_value:
                max_value = value
                max_ingredient = i
    current = current.set_value(index, 'INGREDIENT', max_ingredient)
    
    for index,val in enumerate(current.itertuples()):
        current.set_value(index, 'PROTEINVALUE',float(df1[df1['INGREDIENT'] == val[2]]['PROTEINVALUE']))
    print(current)
    print('total protein value of recipe:',current[["FACTOR",'PROTEINVALUE']].product(axis=1).sum())
    print('\n')
```
